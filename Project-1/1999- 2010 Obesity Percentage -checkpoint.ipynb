```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

obesity_path = "Obesity_among_children.csv"

obesity_data = pd.read_csv(obesity_path, encoding = "UTF-8")

obesity_data
```

```python
obesity_data.columns
```

```python
summarized_obesity =obesity_data.loc[:, ["YEAR", "STUB_LABEL", "ESTIMATE"]]

summarized_obesity.head(30)
```

```python
summarized_obesity_male =summarized_obesity.loc[summarized_obesity["STUB_LABEL"] == 'Male']

summarized_obesity_male.head(10)
```

```python
summarized_obesity_male_Year = summarized_obesity_male.loc[(summarized_obesity_male["YEAR"] == '1999-2002')|
                                                           (summarized_obesity_male["YEAR"] =='2001-2004')|
                                                           (summarized_obesity_male["YEAR"] =='2003-2006')|
                                                            (summarized_obesity_male["YEAR"] =='2005-2008')|
                                                            (summarized_obesity_male["YEAR"] =='2007-2010')]

summarized_obesity_male_Year.head()
```

```python
x_values = summarized_obesity_male_Year["YEAR"]
y_values = summarized_obesity_male_Year["ESTIMATE"]
labels = ["1999-2002", "2002-2004", "2004-2006", "2006-2008", "2008- 2010"]
plt.bar(x_values, y_values)
plt.xlabel("Year")
plt.xticks(x_values, labels, rotation='vertical')
plt.ylabel("Obesity Percentage")
plt.title("Obesity Percentage for Males (Ages 2-19)")
plt.savefig("Images of male")
plt.show()
```

```python
summarized_obesity_female =summarized_obesity.loc[summarized_obesity["STUB_LABEL"] == 'Female']

summarized_obesity_female.head(7)
```

```python
summarized_obesity_female_Year = summarized_obesity_female.loc[(summarized_obesity_female["YEAR"] == '1999-2002')|
                                                           (summarized_obesity_female["YEAR"] =='2001-2004')|
                                                           (summarized_obesity_female["YEAR"] =='2003-2006')|
                                                            (summarized_obesity_female["YEAR"] =='2005-2008')|
                                                            (summarized_obesity_female["YEAR"] =='2007-2010')]

summarized_obesity_female_Year.head()
```

```python
x_values = summarized_obesity_female_Year["YEAR"]
y_values = summarized_obesity_female_Year["ESTIMATE"]
labels = ["1999-2002", "2002-2004", "2004-2006", "2006-2008", "2008- 2010"]
plt.bar(x_values, y_values)
plt.xlabel("Year")
plt.xticks(x_values, labels, rotation='vertical')
plt.ylabel("Obesity Percentage")
plt.title("Obesity Percentage for Females (Ages 2-19)")
plt.savefig("Images of Female")
plt.show()
```

```python
summarized_obesity_overall =summarized_obesity.loc[summarized_obesity["STUB_LABEL"] == '2-19 years']

summarized_obesity_overall.head(10)
```

```python
x_values = summarized_obesity_overall["YEAR"]
y_values = summarized_obesity_overall["ESTIMATE"]

plt.rc('lines', linewidth=1, linestyle='-', marker='o')
labels=["1988-1994","1999-2002","2002-2004","2004-2006","2006-2008","2008-2010","20010-2012","2012-2014",
        "2014-2016","2016-2018"]
plt.plot(x_values, y_values)
plt.xticks(x_values, labels, rotation='vertical')
plt.xlabel("Years")
plt.ylabel("Obesity Percentage")
plt.title("Percent of Children Obese (Ages 2-19)")
plt.savefig("Image of overall")
plt.show()
```
