```python
# select data for summary
panel = "panel_num=1"
total= "&stub_name=Total"

child_totals = requests.get(c_url + panel + total).json()

child_obesity = pd.DataFrame(child_totals)
child_obesity
```

```python
# set up Dataframe to explore obesity with poverty levels
w_poverty = "&stub_name='Percent of poverty level'"
c_obesity_w_poverty = requests.get(c_url + panel + w_poverty).json()

child_obesity_with_poverty = pd.DataFrame(c_obesity_w_poverty)
child_obesity_with_poverty 
```

```python
# clean dataframe columns
child_obesity_with_poverty = child_obesity_with_poverty[["year", "year_num", "age", "stub_label", "estimate"]]
child_obesity_with_poverty 
```

```python
# set estimate as a numeric value
child_obesity_with_poverty["estimate"] = pd.to_numeric(child_obesity_with_poverty["estimate"])
```

```python
# pivot table to create columns based on poverty level
child_obesity_by_poverty_level = pd.DataFrame(child_obesity_with_poverty.pivot_table(index = "year", columns = "stub_label", values = "estimate"))
child_obesity_by_poverty_level.columns
```

```python
# reorder columns
child_obesity_by_poverty_level = child_obesity_by_poverty_level[['Below 100%', '100%-199%', '200%-399%', '400% or more']]
child_obesity_by_poverty_level
```

```python
# export to CSV for double bar graph
child_obesity_by_poverty_level.to_csv("Resources/child_obesity_by_poverty_level_percentage.csv")
```

```python
total = "stratification1=Total"
y_obesity_df = pd.DataFrame(requests.get(y_url + total + "&locationabbr=US" ).json())
y_obesity_df.head()
```

```python
y_obesity_df = y_obesity_df[['yearstart', 'question', 'data_value']]
y_obesity_df.dtypes
```

```python
y_obesity_df['data_value'] = pd.to_numeric(y_obesity_df['data_value'])
y_obesity_df.dtypes
```

```python
#rename column for data clarity
y_obesity_df = y_obesity_df.rename(columns = {'yearstart':'Year'})
```

```python
# pivot table to capture all questions 
y_obesity_factors_df = y_obesity_df.pivot_table(index = 'Year', columns = 'question', values = 'data_value')
y_obesity_factors_df
```

```python
# verify all factors pivoted to columns in the new table
y_obesity_factors_df.columns
```

```python
# clean dataframe (removed "overweight" category as we wanted to focus on obesity)
y_obesity_factors_df = y_obesity_factors_df[['Percent of students in grades 9-12 watching 3 or more hours of television each school day',
       'Percent of students in grades 9-12 who achieve 1 hour or more of moderate-and/or vigorous-intensity physical activity daily',
       'Percent of students in grades 9-12 who consume fruit less than 1 time daily',
       'Percent of students in grades 9-12 who consume vegetables less than 1 time daily',
       'Percent of students in grades 9-12 who drank regular soda/pop at least one time per day',
       'Percent of students in grades 9-12 who have obesity',
       'Percent of students in grades 9-12 who participate in daily physical education']]
y_obesity_factors_df
```

```python
# plot factors on line graph and print to resources
y_obesity_factors_df.plot()
plt.legend(bbox_to_anchor = (1.8, -0.15))

plt.savefig("Images/factors_in_obesity.png")
```

```python
# pull total per state for all years in data set
y_s_dfs = []


for location in locations:
    try:
        y_obesity_by_state = requests.get(y_url + total + Q038 + "&locationdesc=" + location).json()
        
        y_s_dfs.append(pd.DataFrame(y_obesity_by_state))
        
    except: print("no data")
        
        
```

```python
# compile dataframes for all states into one
y_obesity_state_df=pd.concat(y_s_dfs)
y_obesity_state_df.head()
```

```python

youth_obesity_state_sum = y_obesity_state_df[["yearstart","locationdesc", "data_value"]]
youth_obesity_state_sum = youth_obesity_state_sum.rename(columns = {"yearstart": "Year", "locationdesc": "Location", "data_value": "% Obesity in Adolescents"})

youth_obesity_state_sum.head()
```

```python
# pull adult data (to see if policy had long-term effects on lifestyles)
a_dfs = []
Q036="questionid=Q036"
total = "stratification1=Total"

for location in locations:
    try:
        a_obesity_data = requests.get(a_url + Q036 + "&" + total + "&locationdesc=" + location).json()
        
        a_dfs.append(pd.DataFrame(a_obesity_data))
        
    except: print("no data")
        
        
```

```python
# capture data for states
a_obesity_df=pd.concat(a_dfs)
a_obesity_df.head()
```

```python
# pull state values over years to dataframe for possible chart analysis
adult_obesity_sum = a_obesity_df[["locationdesc", "yearstart", "data_value"]]

adult_obesity_sum["data_value"] = pd.to_numeric(adult_obesity_sum["data_value"])
adult_obesity_summary = pd.DataFrame(adult_obesity_sum.pivot_table(index = "locationdesc", columns = "yearstart", values = "data_value"))
```
