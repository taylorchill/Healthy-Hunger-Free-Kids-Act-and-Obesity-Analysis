```python
%matplotlib notebook

import pandas as pd
from matplotlib import pyplot as plt

import numpy as np
import scipy.stats as st
from scipy.stats import linregress
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hthelander/Project-1/clean_data/Resources/youth_obesity_state_scatter.csv")
df
```

```python
df.columns
```

```python
df['Year'].value_counts()
```

```python


# Set x axis and tick locations

x_axis = np.arange(2001,2020)
tick_locations = [value+0.4 for value in x_axis]

plt.scatter(df["Year"], df["% Obesity in Adolescents"], color='b', alpha=0.5)

plt.xticks(tick_locations, x_axis,rotation =90)
plt.savefig('Images/scatterplot_with_equation')
plt.show()








```

```python
x_values = df['Year']
y_values = df['% Obesity in Adolescents']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")

plt.annotate(line_eq,(2002,18),fontsize=15,color="red")

plt.xlabel('Year')
plt.ylabel('% Obesity')
print(f"The r-squared is: {rvalue**2}")

plt.show()
# Save the figure.
plt.savefig('scatterplot_1')
```
